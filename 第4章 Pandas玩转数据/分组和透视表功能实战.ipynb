# 3.每个航空公司延误的情况

```python
delay_group = df.groupby(['unique_carrier','delayed'])
```

```python
delay_group
```

```python
delay_group.size()
```

```python
df_delay = delay_group.size().unstack()
```

```python
# 把多级Series转化成DataFrame
df_delay
```

```python
import matplotlib.pyplot as plt
```

```python
df_delay.plot()
```

```python
plt.show()
```

```python
df_delay.plot(kind='barh',stacked=True,figsize=[16,6],colormap='winter')
```
