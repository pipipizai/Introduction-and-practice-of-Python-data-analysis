```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
%matplotlib inline
```

```python
import seaborn as sns
```

```python
s1 = Series(np.random.randn(1000))
```

```python
plt.hist(s1)
```

```python
s1.plot(kind='kde')
```

```python
sns.distplot(s1,hist=True,kde=True)
```

```python
sns.distplot(s1,hist=True,kde=False)
```

```python
sns.distplot(s1,hist=True,kde=True,rug=True)
```

```python
sns.distplot(s1,bins=20,hist=True,kde=False,rug=True)
```

```python
sns.kdeplot(s1)
```

```python
sns.kdeplot(s1,shade=True)
```

```python
sns.kdeplot(s1,shade=True,color='r')
```
