# 1.axes_style and set_style

```python
x = np.linspace(0,14,100)
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
def sinplot():
    plt.plot(x,y1)
    plt.plot(x,y2)
```

```python
sinplot()
```

```python
import seaborn as sns
```

```python
sinplot()
```

```python
style = ['darkgrid','dark','white','whitegrid','tricks']
```

```python
sns.set_style(style[0])
sinplot()
```

```python
sns.set_style(style[1])
sinplot()
```

```python
sns.set_style(style[2])
sinplot()
```

```python
sns.set_style(style[0])
sinplot()
sns.axes_style()
```

```python
sns.set_style(style[0],{'grid.color':'red'})
sinplot()
sns.axes_style()
```

```python
sinplot()
```

```python
#
sns.set()
```

```python
sns.set_style('white')
sinplot()
```

# 2.plotting_context() and set_context()
